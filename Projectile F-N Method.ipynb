The Feynman-Newton method is a method of solving problems in classical mechanics using the path integral formulation of quantum mechanics. The method can be used to study the path of a projectile, with and without air drag, by treating the problem as a quantum mechanical system.

```python
import numpy as np
from scipy.integrate import odeint

# Constants
m = 1.0 # mass of projectile
g = 9.8 # acceleration due to gravity
b = 0.1 # coefficient of air drag

# Initial conditions
x0 = 0.0 # initial horizontal position
y0 = 0.0 # initial vertical position
vx0 = 10.0 # initial horizontal velocity
vy0 = 10.0 # initial vertical velocity

# Define the right-hand side of the ODE
def rhs(y, t, m, g, b):
    x, y, vx, vy = y
    dxdt = vx
    dydt = vy
    dvxdt = -b/m*vx
    dvydt = -g - b/m*vy
    return [dxdt, dydt, dvxdt, dvydt]

# Time grid for integration
t = np.linspace(0, 10, 100)

# Solve the ODE with air drag
y = odeint(rhs, [x0, y0, vx0, vy0], t, args=(m, g, b))
x = y[:,0]
y = y[:,1]

# Plot the solution
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
