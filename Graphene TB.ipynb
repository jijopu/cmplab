```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 2.46 # lattice constant
t = -2.8 # hopping parameter
N = 100 # number of lattice sites
kx = np.linspace(-np.pi/a, np.pi/a, N)

# Create the Hamiltonian
H = np.zeros((N,N))
for i in range(N-1):
    H[i,i+1] = t
    H[i+1,i] = t

# Add the nearest-neighbor hopping term
for i in range(N-3):
    H[i,i+3] = t
    H[i+3,i] = t

# Add the second-nearest-neighbor hopping term
for i in range(N-4):
    H[i,i+4] = t
    H[i+4,i] = t

# Find the eigenvalues and eigenvectors
eigenvalues = []
for k in range(N):
    H_k = H + 2 * t * np.cos(kx[k] * a) * np.eye(N)
    eigenvalues.append(np.linalg.eigvals(H_k))

# Plot the band structure
eigenvalues = np.array(eigenvalues).T
for band in eigenvalues:
    plt.plot(kx, band)
plt.xlabel('kx')
plt.ylabel('Energy (eV)')
plt.title('Band Structure of Graphene')
plt.show()
```
