## Damped Driven Harmonic Oscillator
The equation of a damped driven harmonic oscillator is given below. Given the following initial conditions, at initial time, position $x = 0.2$ and velocity $= 0.8$, $b = 0.2$ and $A = 0.1$. Resonant frequency $= 1.2$ and driving frequency $1$.


$\ddot{x}=-ax-b\dot{x}+A\sin(\omega_0 t)$

### Time Series

```python
import numpy as np
import matplotlib.pyplot as plt

#initial values
x=0.2
xdot=0.8
b=0.2
A=0.2
a=(1.2)**2     #a=(Resonant frequency)**2

t=0
h=0.01

DriF=1

E=100


def Xddot(x,xdot,t):
    return (-a*x)-(b*xdot)+(A*np.sin(DriF*t))

X=[x]
T=[t]
Xdot=[xdot]
# Integration using Euler method
while t<E:
    x=x+(h*xdot)
    t=t+h
    xdot=xdot+(Xddot(x,xdot,t)*h)
    X=X+[x]
    T=T+[t]
    Xdot=Xdot+[xdot]

plt.figure(figsize=(10,5), dpi=100)
plt.title("Damped Driven Harmonic Oscillator")
plt.xlabel("t (Time)")
plt.ylabel("x (Position)")
plt.plot(T,X)
plt.grid()
```

### Phase Plot

```python
plt.figure(figsize=(10,5), dpi=100)
plt.title("Damped Driven Harmonic Oscillator - Phase Plot")
plt.xlabel("x (Position)")
plt.ylabel("v (Velocity)")
plt.plot(X,Xdot)
plt.grid()
```

### Energy vs Time

```python
X=np.array(X)
T=np.array(T)
Xdot=np.array(Xdot)

plt.figure(figsize=(10,5), dpi=100)
plt.title("Damped Driven Harmonic Oscillator  E/m vs t")
plt.ylabel("E_m (Energy parameter)")
plt.xlabel("t (Time)")
m=1                                 #mass
E_m=(Xdot**2/2)+(a*m*X**2/2)        #E_m=(v^2/2)+((k/m)^2*x^2/2)  i.e. (KE/m)+(PE/m)
plt.plot(T,E_m)
```
