## Lorenz Attractor

$\frac{dx}{dt}=\sigma(y-x)$

$\frac{dy}{dt}=x(\rho-z)-y$

$\frac{dz}{dt}=(xy-\beta z)$

The parameters are $\sigma = 10$,  $\beta = 8/3$,  $\rho = 28$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s=10                      #sigma
b=8/3                     #beta
r=28                      #rho

t=0                       #initial time
h=0.001                   #time step
E=100                     #maximum time

x=y=z=0.1                 #initial x,y,z

def fx(x,y):              #dx/dt
    return s*(y-x)

def fy(x,y,z):            #dy/dt
    return x*(r-z)-y

def fz(x,y,z):            #dz/dt
    return (x*y)-(b*z)

X=[x]
Y=[y]
Z=[z]
T=[t]
while t<=E:
    xi=x
    yi=y
    zi=z
    x=x+fx(xi,yi)*h
    y=y+fy(xi,yi,zi)*h
    z=z+fz(xi,yi,zi)*h
    X+=[x]
    Y+=[y]
    Z+=[z]
    t=t+h
    T=T+[t]
```

```python
# x-z plot
plt.figure(figsize=(10,5),dpi=200)
plt.axes(facecolor="black")
plt.xlabel("x")
plt.ylabel("z")
plt.plot(X,Z,'#862696',linewidth=0.5)
plt.show()
```

```python
# x-y plot
plt.figure(figsize=(10,5),dpi=200)
plt.axes(facecolor="black")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(X,Y,'brown',linewidth=0.5)
plt.show()
```

```python
# y-z plot
plt.figure(figsize=(10,5),dpi=200)
plt.axes(facecolor="black")
plt.xlabel("y")
plt.ylabel("z")
plt.plot(Y,Z,'skyblue',linewidth=0.5)
plt.show()
```
