# Lauguerre Polynomial

$(n+1)L_n(x)=(2n+1-x)L_{(n-1)}(x)-(n-2)L_{(n-2)}(x)$

```python
def laug(x,n):
    if n==0:
        return 1
   # elif n==1:
  #      return 2*x
    else:
        return (2*n+1-x)*laug(x,n-1)-(n-2)*laug(x,n-2)
```

```python
for i in range(0,5):
    print(laug(1,i))
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
def HERMITE(X,N):
    HER = hermite(N)
    sn  = HER(X)
    return sn
xvals = np.linspace(-1.0,1.0,1000)
for n in np.arange(0,2,1):
    sol = HERMITE(xvals,n)
    plt.plot(xvals,sol,"-.",label = "n = " + str(n),linewidth=2)
plt.xticks(fontsize=14,fontweight="bold")
plt.yticks(fontsize=14,fontweight="bold")
plt.grid()
plt.legend()
plt.show()
```
