```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
mu_0 = 4*np.pi*10**-7  # Vacuum permeability

# Coil parameters
N = 100  # Number of turns
I = 1  # Current
R = 0.1  # Radius

# Simulation parameters
x = np.linspace(-1, 1, 100)  # Distance along the axis

# Magnetic field calculation
B = (mu_0*N*I)/(2*np.pi*R) * (x/np.sqrt(R**2 + x**2))

# Plotting
plt.plot(x, B)
plt.xlabel('Distance along the axis (m)')
plt.ylabel('Magnetic field (T)')
plt.title('Magnetic field along the axis of a coil')
plt.show()
```
