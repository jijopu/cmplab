Bose-Einstein distribution and uses it to prove the Stefan-Boltzmann law and Wein's displacement law

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k_B = 1.38e-23  # Boltzmann constant (J/K)
h = 6.626e-34  # Planck constant (J*s)
c = 3e8  # Speed of light (m/s)

# Temperature range (K)
T_min = 1e-3
T_max = 1e3
T_range = np.linspace(T_min, T_max, 100)

# Bose-Einstein distribution
def bose_einstein(E, T):
    return 1/(np.exp(E/(k_B*T)) - 1)

# Plot the Bose-Einstein distribution
E_range = np.linspace(0, 10*k_B*T_max, 1000)
for T in T_range:
    plt.plot(E_range/(k_B*T), bose_einstein(E_range, T), label=f'T = {T:.1f} K')
plt.xlabel('Energy (kT)')
plt.ylabel('Occupation number')
plt.legend()
plt.show()
```

```python
# Proving Stefan-Boltzmann law

# Stefan-Boltzmann constant
sigma = (np.pi**2*k_B**4)/(60*h**3*c**2)

# Stefan-Boltzmann law
def stefan_boltzmann(T):
    return sigma*T**4

# Plotting Stefan-Boltzmann law
plt.plot(T_range, stefan_boltzmann(T_range), 'r-', label='Stefan-Boltzmann law')
plt.xlabel('Temperature (K)')
plt.ylabel('Power density (W/m^2)')
plt.legend()
plt.show()
```

```python
# Proving Wein's displacement law

# Wavelength range
lambda_min = 1e-9
lambda_max = 1e-6
lambda_range = np.linspace(lambda_min, lambda_max, 100)

# Wein's displacement law
def wein(T, wavelength):
    return 2*h*c**2/(wavelength**5*(np.exp(h*c/(wavelength*k_B*T)) - 1))

# Plotting Wein's displacement law
for T in T_range:
    plt.plot(lambda_range*1e9, wein(T, lambda_range), label=f'T = {T:.1f} K')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Power density (W/m^3)')
plt.legend()
plt.show()
```
